```python
!pwd
%cd /content/gdrive/My Drive/MLGT/PA2/results/maddpg_uav
!ls

file = open('test_agrewards.pkl', 'rb')
uav = pickle.load(file)
print(uav)
print(len(uav))

import matplotlib.pyplot as plt
length_x = len(uav) // 3
x = [500 * (1+i) for i in range(length_x)]


fig, ax = plt.subplots()
ax.plot(x, uav[0::3], 'r', label='poacher')
ax.plot(x, uav[1::3], 'b', label='ranger')
ax.plot(x, uav[2::3], 'g', label='uav')
plt.xlabel('number of episodes')
plt.ylabel('agent\'s reward')
leg = ax.legend()
```
